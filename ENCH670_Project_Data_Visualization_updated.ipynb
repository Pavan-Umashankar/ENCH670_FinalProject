```python
import pandas as pd
import matplotlib.pyplot as plt

Yeast_Raw = pd.read_csv('ENCH470_YeastDataSet - UpdatedDataSet_CSV.csv')
YeastData = Yeast_Raw.drop('Unnamed: 7', axis='columns')
YeastData
```

```python
# Scatter plots of raw data

Yeast_glu = YeastData[YeastData['Glucose']==1]
Yeast_mal = YeastData[YeastData['Maltodextrin']==1]

fig, ax = plt.subplots(2,2, figsize=(12,6))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

ax[0,0].scatter(Yeast_glu['T (C)'],Yeast_glu['mu-max (1/hr)'],alpha=0.5)
ax[0,0].scatter(Yeast_mal['T (C)'],Yeast_mal['mu-max (1/hr)'],alpha=0.5)
ax[0,0].set_xlabel(r'Temperature ($\circ$C)')
ax[0,0].set_ylabel(r'$\mu$ max (1/hr)')
ax[0,0].legend(['glucose','maltodextrin'])

ax[0,1].scatter(Yeast_glu['pH'],Yeast_glu['mu-max (1/hr)'],alpha=0.5)
ax[0,1].scatter(Yeast_mal['pH'],Yeast_mal['mu-max (1/hr)'],alpha=0.5)
ax[0,1].set_xlabel('pH')
ax[0,1].set_ylabel(r'$\mu$ max (1/hr)')
ax[0,1].legend(['glucose','maltodextrin'],loc = 'lower right')

ax[1,0].scatter(Yeast_glu['DO (%)'],Yeast_glu['mu-max (1/hr)'],alpha=0.5)
ax[1,0].scatter(Yeast_mal['DO (%)'],Yeast_mal['mu-max (1/hr)'],alpha=0.5)
ax[1,0].set_xlabel('Dissolved Oxygen (%)')
ax[1,0].set_ylabel(r'$\mu$ max (1/hr)')
ax[1,0].legend(['glucose','maltodextrin'])

ax[1,1].scatter(Yeast_glu['Sugar Concentration (g/L)'],Yeast_glu['mu-max (1/hr)'],alpha=0.5)
ax[1,1].scatter(Yeast_mal['Sugar Concentration (g/L)'],Yeast_mal['mu-max (1/hr)'],alpha=0.5)
ax[1,1].set_xlabel('Sugar Concentration (g/L)')
ax[1,1].set_ylabel(r'$\mu$ max (1/hr)')
ax[1,1].legend(['glucose','maltodextrin'])

plt.show()
```

```python
# Vary Temperature, others constant

Yeast_temp = YeastData.drop(YeastData.index[13:47])
Yeast_temp

fig, ax = plt.subplots()
ax.scatter(Yeast_temp['T (C)'],Yeast_temp['mu-max (1/hr)'],alpha=0.5)
ax.set_xlabel('Temperature (C)')
ax.set_ylabel(r'$\mu$ max (1/hr)')

plt.show()
```

```python
# Vary DO %, others constant
# Broke down dataframe into 3 (pH 4,5,6)

Yeast_DO_4 = YeastData.iloc[13:16,:]
Yeast_DO_5 = YeastData.iloc[16:19,:]
Yeast_DO_6 = YeastData.iloc[19:22,:]

fig, ax = plt.subplots()
ax.scatter(Yeast_DO_4['DO (%)'],Yeast_DO_4['mu-max (1/hr)'],alpha=0.5)
ax.scatter(Yeast_DO_5['DO (%)'],Yeast_DO_5['mu-max (1/hr)'],alpha=0.5)
ax.scatter(Yeast_DO_6['DO (%)'],Yeast_DO_6['mu-max (1/hr)'],alpha=0.5)
ax.set_xlabel('Dissolved Oxygen (%)')
ax.set_ylabel(r'$\mu$ max (1/hr)')
ax.legend(['pH 4','pH 5','pH 6'])

plt.show()
```

```python
# Vary sugar con, others constant

Yeast_SC_28 = YeastData.iloc[22:27,:]
Yeast_SC_30 = YeastData.iloc[[32,36,40,44],:]

fig, ax = plt.subplots(2,1, figsize=(6,6))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

ax[0].scatter(Yeast_SC_28['Sugar Concentration (g/L)'],Yeast_SC_28['mu-max (1/hr)'],alpha=0.5)
ax[0].set_xlabel('Sugar Concentration (g/L)')
ax[0].set_ylabel(r'$\mu$ max (1/hr)')
ax[0].legend(['28C'])

ax[1].scatter(Yeast_SC_30['Sugar Concentration (g/L)'],Yeast_SC_30['mu-max (1/hr)'],alpha=0.5)
ax[1].set_xlabel('Sugar Concentration (g/L)')
ax[1].set_ylabel(r'$\mu$ max (1/hr)')
ax[1].legend(['30C'])

plt.show()
```
